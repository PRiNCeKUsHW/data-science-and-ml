```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

NOte : seaborn has some built in dataset

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting the theme for plots
sns.set_theme()
```

```python
# visulize the data
sns.relplot(data=tips,x = 'total_bill',y = 'tip', col='time',hue='smoker',style='smoker',size='size')
```

```python
# house price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()

house = pd.DataFrame(house_boston.data,columns=house_boston.feature_names)
house['PRICE'] = house_boston.target
```

```python
print(house_boston)
```

```python
house.head()
```

distribution plot


```python
sns.distplot(house['PRICE'])
```

correlation matrix

-postitive

-negative

Heat map

```python
corretlation = house.corr()
```

```python
# constructing heat map
plt.figure(figsize=(10,10))
sns.heatmap(corretlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Oranges')
```
