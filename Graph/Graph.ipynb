```python
x = np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="black",linewidth=3,label="x^3")
plt.plot(x,y2,color="blue",linewidth=3,label="x^2")
plt.legend()
plt.ylabel("y")

plt.xlabel("x")
plt.title("matplot")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black')
plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black',marker='^')
plt.title("bubble chart")
plt.show()
```

```python
c = np.arange(len(year))
a = len(c)
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black',marker='^')
plt.title("bubble chart")
for i in range(a):
    plt.text(year[i]+0.1,salary[i]+0.1,population[i])
plt.show()
```

```python
def func(sold,company):
    li=[]
    sum1 = sum(sold)
    for i in range(len(sold)):
        num = (sold[i]*100)/sum1
        print(company[i],round(num,1))
        li.append(num)
    s=str(li)
    return s

company = ['hp','dell','lenovo','asus','appale','acer']
sold = [20000,43000,15000,17000,22000,13000]
plt.pie(sold,labels=company,explode=[0.1,0.2,0.1,0.2,0.1,0],autopct = func(sold,company),counterclock=False,startangle=90)
plt.show()
```
